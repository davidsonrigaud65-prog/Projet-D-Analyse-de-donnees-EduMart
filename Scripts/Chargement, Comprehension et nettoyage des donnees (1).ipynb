# Etape 2 : Nettoyage des donnees


```python
# On remplace 'unknown' par NaN
customers['age'] = customers['age'].replace('unknown', np.nan)

#  puis on convertit en float
customers['age'] = pd.to_numeric(customers['age'], errors='coerce')

# On remplace les valeurs manquantes par la mediane
age_mediane = customers['age'].median()
customers['age'] = customers['age'].fillna(age_mediane)
```

```python
# Pour verification des ages 
print(customers['age'].unique())
```

```python
# Gestion des Dates 
customers['signup_date'] = pd.to_datetime(customers['signup_date'])
orders['order_date'] = pd.to_datetime(orders['order_date'])
```

```python
#  Conversion de discount_pct en format numeric
orders['discount_pct'] = pd.to_numeric(orders['discount_pct'], errors='coerce')
```

```python
# Traitement des autres valeurs manquantes 

customers['city'] = customers['city'].fillna('Inconnu')
customers['gender'] = customers['gender'].fillna('Inconnu')

# On remplace le ccore moyen pour les avis manquants et 0 pour les délais
orders['review_score'] = orders['review_score'].fillna(orders['review_score'].mean())
orders['delivery_days'] = orders['delivery_days'].fillna(0)
```

```python
# Constats sur les doublons avant modifications
nb_doublons = orders.duplicated().sum()

print(nb_doublons)

print(orders[orders.duplicated()].head())
```

```python
# Suppression des Doublons et Anomalies

#  Supprimer les doublons parfaits dans la table des ventes
orders = orders.drop_duplicates()

#  Supprimer les quantités négatives ou nulles
orders = orders[orders['quantity'] > 0]

# Supprimer les délais de livraison (> 30 jours)
orders = orders[orders['delivery_days'] <= 30]
```

```python
# Vérification de la cohérence financière
# Calcul théorique
net_theorique = orders['gross_amount'] * (1 - orders['discount_pct'])

print(orders['net_amount'])
print(net_theorique)
```

```python
# On ne garde que les lignes où l'écart est presque nul (tolérance de 0.01)
ecart = np.abs(orders['net_amount'] - net_theorique)
orders = orders[ecart < 0.01]
```

# Les 5 règles de nettoyage documentées

###
1.- Normalisation des Types Temporels : 
Conversion des colonnes order_date et signup_date en format datetime pour permettre les analyses chronologiques précises.

2.- Traitement des Valeurs Textuelles dans l'Âge : 
Dans la table customers, les entrées "unknown" ont été converties en valeurs numériques NaN. Les valeurs manquantes ont ensuite été complétées par la médiane (28 ans) pour ne pas fausser la moyenne d'âge.

3.- Élimination des Doublons Transactionnels : 
Nous avons identifié et supprimé 25 lignes identiques dans order_lines. Cette règle garantit que le chiffre d'affaires et le nombre de commandes ne sont pas surestimés.

4.- Filtrage des Anomalies de Saisie : 
Suppression de toutes les transactions avec une quantity inférieure ou égale à 0 (erreurs de stock) et des livraisons supérieures à 30 jours (délais logistiques jugés aberrants pour EduMart).

5.- Audit de Cohérence Financière : 
Recalcul du montant net pour chaque ligne en fonction du prix brut et de la remise. Seules les lignes où l'écart est inférieur à 0.01 ont été conservées pour assurer la fiabilité des futurs calculs de profitabilité.
